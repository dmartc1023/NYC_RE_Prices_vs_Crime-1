```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
#import numpy as np
#import dateutil

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
nyc_rental_load = "\\Users\\Paul-DS\\Downloads\\Neighborhood_Zri_AllHomesPlusMultifamily.csv"
nyc_rental_data=pd.read_csv(nyc_rental_load)
#nyc_mean_price_sales = "Resources/nyc_mean_price_per_sqft.csv"
#nyc_price_per_sqft = pd.read_csv(nyc_mean_price_sales)
```

```python
manhattan_rent = nyc_rental_data[nyc_rental_data["CountyName"]=="New York"]
```

```python
manhattan_rent_mean = manhattan_rent.drop(columns=["RegionID","RegionName","City","State","Metro","SizeRank","CountyName"]).mean()
```

```python
bronx_rent = nyc_rental_data[nyc_rental_data["CountyName"]=="Bronx"]
bronx_rent_mean = manhattan_rent.drop(columns=["RegionID","RegionName","City","State","Metro","SizeRank","CountyName"]).mean()
```

```python
queens_rent = nyc_rental_data[nyc_rental_data["CountyName"]=="Queens"]
queens_rent_mean = manhattan_rent.drop(columns=["RegionID","RegionName","City","State","Metro","SizeRank","CountyName"]).mean()
```

```python
staten_island_rent = nyc_rental_data[nyc_rental_data["CountyName"]=="Richmond"]
staten_island_rent_mean = manhattan_rent.drop(columns=["RegionID","RegionName","City","State","Metro","SizeRank","CountyName"]).mean()
```

```python
brooklyn_rent = nyc_rental_data[nyc_rental_data["CountyName"]=="Kings"]
brooklyn_rent_mean = manhattan_rent.drop(columns=["RegionID","RegionName","City","State","Metro","SizeRank","CountyName"]).mean()
```

```python
print(manhattan_rent_mean.head())
print(bronx_rent_mean.head())
print(queens_rent_mean.head())
print(staten_island_rent_mean.head())
print(brooklyn_rent_mean.head())
```

```python
df = pd.DataFrame(manhattan_rent_mean,columns=['Manhattan rental price'])
```

```python
df = df.join(pd.DataFrame(brooklyn_rent_mean,columns=['Brooklyn rental price']))
```

```python
df = df.join(pd.DataFrame(bronx_rent_mean,columns=['Bronx rental price']))
```

```python
df = df.join(pd.DataFrame(queens_rent_mean, columns=['Queens rental price']))
```

```python
df = df.join(pd.DataFrame(staten_island_rent_mean, columns = ['Staten Island rental price']))
```

```python
df = df.reset_index(drop=False)
```
