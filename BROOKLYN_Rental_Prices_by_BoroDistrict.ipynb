```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
#import numpy as np
#import dateutil

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
nyc_rental_load = "\\Users\\Paul-DS\\Downloads\\Neighborhood_Zri_AllHomesPlusMultifamily.csv"
nyc_rental_data=pd.read_csv(nyc_rental_load)
#nyc_mean_price_sales = "Resources/nyc_mean_price_per_sqft.csv"
#nyc_price_per_sqft = pd.read_csv(nyc_mean_price_sales)
```

```python
brooklyn_rent = nyc_rental_data[nyc_rental_data["CountyName"]=="Kings"]
```

```python
brooklyn_region=[]
brooklyn_region = brooklyn_rent['RegionName'].unique()
```

```python
brooklyn_region
```

```python
nyc_region_rent={}
for region in brooklyn_region:
    nyc_region_rent[region] = brooklyn_rent[nyc_rental_data["RegionName"]==region]
    nyc_region_rent[region] = nyc_region_rent[region].drop(columns=["RegionID","RegionName","City","State","Metro","SizeRank","CountyName"]).mean()
    nyc_region_rent[region] = pd.DataFrame(nyc_region_rent[region],columns=[region+' rental price'])
```

```python
#nyc_region_rent
for region in brooklyn_region:
    nyc_region_rent[region].index.rename('Date',inplace=True)
    #manhattan_region_df = manhattan_region_df.concat(nyc_region_rent[region])
    #manhattan_region_df = manhattan_region_df.join(nyc_region_rent[region])
#nyc_region_rent
```

```python
first=0
for region in brooklyn_region:
    if first==0:
        df = pd.DataFrame(nyc_region_rent[region])
    else:
        df = df.join(pd.DataFrame(nyc_region_rent[region]))
    first=1
```
