```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
#import numpy as np
#import dateutil

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
nyc_price_load = "\\Users\\Paul-DS\\Downloads\\Neighborhood_MedianValuePerSqft_AllHomes.csv"
nyc_price_data=pd.read_csv(nyc_price_load)
#nyc_mean_price_sales = "Resources/nyc_mean_price_per_sqft.csv"
#nyc_price_per_sqft = pd.read_csv(nyc_mean_price_sales)
```

```python
manhattan_price = nyc_price_data[nyc_price_data["CountyName"]=="New York"]
```

```python
manhattan_region=[]
manhattan_region = manhattan_price['RegionName'].unique()
```

```python
manhattan_region
```

```python
nyc_region_price={}
for region in manhattan_region:
    nyc_region_price[region] = manhattan_price[manhattan_price["RegionName"]==region]
    nyc_region_price[region] = nyc_region_price[region].drop(columns=["RegionID","RegionName","City","State",
                                                                      "Metro","SizeRank","CountyName"]).mean()
    nyc_region_price[region] = pd.DataFrame(nyc_region_price[region],columns=[region+' Median Price Sqft'])
```

```python
nyc_region_price[region].index.rename('Date')
```

```python
#nyc_region_rent
for region in manhattan_region:
    nyc_region_price[region].index.rename('Date',inplace=True)
    #manhattan_region_df = manhattan_region_df.concat(nyc_region_rent[region])
    #manhattan_region_df = manhattan_region_df.join(nyc_region_rent[region])
#nyc_region_rent
```

```python
first=0
for region in manhattan_region:
    if first==0:
        df = pd.DataFrame(nyc_region_price[region])
    else:
        df = df.join(pd.DataFrame(nyc_region_price[region]))
    first=1
```
