```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import dateutil

# this code written by Paul AG
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
nyc_zillow_load = "\\Users\\Paul-DS\\Downloads\\Neighborhood_MedianValuePerSqft_AllHomes.csv"
nyc_zillow_data=pd.read_csv(nyc_zillow_load)
```

```python
manhattan_val_per_sqrt_foot = nyc_zillow_data[nyc_zillow_data["CountyName"]=="New York"]
```

```python
manhattan_val_per_sqrt_foot_mean = manhattan_val_per_sqrt_foot.drop(columns=["SizeRank","RegionID","RegionName","City","State","Metro","CountyName"]).mean()
```

```python
kings_val_per_sqrt_foot = nyc_zillow_data[nyc_zillow_data["CountyName"]=="Kings"]
```

```python
kings_val_per_sqrt_foot_mean = kings_val_per_sqrt_foot.drop(columns=["SizeRank","RegionID","RegionName","City","State","Metro","CountyName"]).mean()
```

```python
bronx_val_per_sqrt_foot = nyc_zillow_data[nyc_zillow_data["CountyName"]=="Bronx"]
bronx_val_per_sqrt_foot_mean = bronx_val_per_sqrt_foot.drop(columns=["SizeRank","RegionID","RegionName","City","State","Metro","CountyName"]).mean()
```

```python
staten_island_val_per_sqrt_foot = nyc_zillow_data[nyc_zillow_data["CountyName"]=="Richmond"]
staten_island_val_per_sqrt_foot_mean = staten_island_val_per_sqrt_foot.drop(columns=["SizeRank","RegionID","RegionName","City","State","Metro","CountyName"]).mean()
```

```python
queens_val_per_sqrt_foot = nyc_zillow_data[nyc_zillow_data["CountyName"]=="Queens"]
queens_val_per_sqrt_foot_mean = queens_val_per_sqrt_foot.drop(columns=["SizeRank","RegionID","RegionName","City","State","Metro","CountyName"]).mean()
```

```python
print(kings_val_per_sqrt_foot_mean.head())
print(bronx_val_per_sqrt_foot_mean.head())
print(manhattan_val_per_sqrt_foot_mean.head())
print(queens_val_per_sqrt_foot_mean.head())
print(staten_island_val_per_sqrt_foot_mean.head())
```

```python
df = pd.DataFrame(manhattan_val_per_sqrt_foot_mean,columns=['Manhattan price per sqft'])
```

```python
df = df.join(pd.DataFrame(kings_val_per_sqrt_foot_mean,columns=['Brooklyn price per sqft']))
```

```python
df = df.join(pd.DataFrame(bronx_val_per_sqrt_foot_mean,columns=['Bronx price per sqft']))
```

```python
df = df.join(pd.DataFrame(queens_val_per_sqrt_foot_mean, columns=['Queens price per sqft']))
```

```python
df = df.join(pd.DataFrame(staten_island_val_per_sqrt_foot_mean, columns = ['Staten Island price per sqft']))
```

```python
df = df.reset_index(drop=False)
```

```python
df = df.rename(columns={"index":"Date"})
```
